```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
#Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#Set the y-axis to a list of floats as a total fare
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()
#fig = plt.figure()
#ax = fig.add_subplot()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#Creat the plot
#plt.plot(x_axis, y_axis)
#plt.show()
```

```python
#Create the plot and add labels
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0, 45)
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
#Create the plot and add labels
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0, 45)
plt.title("PyBer Fare by Month")
plt.legend()
ax.grid()
```

```python
#Creat a bar plot
plt.bar(x_axis, y_axis)
```

```python
#Create a bar plot
plt.bar(x_axis, y_axis, color="green", label='Boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
#Create a horizontal bar plot/option to switch x and y
plt.barh(x_axis, y_axis)
```

```python
#Create a horizontal bar plot inverted
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
#Create a horizontal bar plot inverted
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
#Create a plot with the ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
#Create a horizontal plot with the ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
#Create a horizontal plot with the ax.plt() inverted
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
ax.invert_yaxis()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
ax.legend()
```

```python
#Create a scatter plot
plt.scatter(y_axis, x_axis, color="red", label='Chicago')
#plt.plot(x_axis, y_axis, 'o')
plt.xlim(0, 50)
plt.xlabel("Fare(&)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.legend()
```

```python
#Create a bubble chart
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Create a bubble chart
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#List comprehension
#new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#Create plot with ax
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Create a horizontal plot with the ax.plt() inverted
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], linewidth=2, color="skyblue", edgecolors="black", alpha=0.8, label='Chicago')
ax.invert_yaxis()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
ax.set_xlim(0, 50)
ax.legend()
```

```python
#Create a pie chart
#plt.pie(y_axis, labels=x_axis)
#plt.show()
```

```python
#Explode Values
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0,)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)

plt.show()
```
